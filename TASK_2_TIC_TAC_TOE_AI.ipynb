```python
import math
import copy
Z=0
X=1
O=-1
Red="\033[91m"
Green="\033[92m"
Reset= "\033[0m"
def board(n):
    for r in n:
        print(" | ".join([
            (Red+"X"+Reset) if s==X else
            (Green+"O"+Reset) if s==O else
            " "
            for s in r
        ]))
        print("-"*25)
def winner(n,p):
    for i in range(3):
        if all([s==p for s in n[i]]) or all([n[j][i]==p for j in range(3)]):
            return True
    if all([n[i][i]==p for i in range(3)]) or all([n[i][2 - i]==p for i in range(3)]):
        return True
    return False
def done(board):
    return any([winner(n,X),winner(n,O), all(s!=Z for r in n for s in r)])
def MinMax(n,d,p,al,be):
    if winner(n,X):
        return -1
    if winner(n,O):
        return 1
    if not any(Z in r for r in n):
        return 0

    if p==O:
        m=-math.inf
        for i in range(3):
            for j in range(3):
                if n[i][j]==Z:
                    n[i][j]=p
                    ans=MinMax(n,d+1,X,al,be)
                    n[i][j]=Z
                    m=max(m,ans)
                    al=max(al,ans)
                    if be<=al:
                        break
        return m
    else:
        m1=math.inf
        for i in range(3):
            for j in range(3):
                if n[i][j]==Z:
                    n[i][j]=p
                    ans=MinMax(n,d+1,O,al,be)
                    n[i][j]=Z
                    m1=min(m1,ans)
                    be=min(be,ans)
                    if be<=al:
                        break
        return m1
def best(n):
    b=-math.inf
    best=None
    for i in range(3):
        for j in range(3):
            if n[i][j]==Z:
                n[i][j]=O
                ans=MinMax(n,0,X,-math.inf,math.inf)
                n[i][j]=Z
                if ans>b:
                    b=ans
                    best=(i, j)
    return best
if __name__ == "__main__":
    n=[[Z for _ in range(3)] for _ in range(3)]

    print("Welcome to Tic-Tac-Toe!")
    board(n)

    while not done(n):
        try:
            while True:
                v=input("Enter your move (1-9): ")
                if v.isdigit():
                    value=int(v)
                    if 1<=value<= 9:
                        r=(value-1)//3
                        c=(value-1)%3
                        if n[r][c]==Z:
                            n[r][c]=X
                            board(n)
                            break
                        else:
                            print("Invalid move.That position is already taken. Try again.")
                    else:
                        print("Invalid input.Please enter a number between 1 and 9.")
                else:
                    print("Invalid input.Please enter a number between 1 and 9.")
        except ValueError:
            print("Invalid input.Please enter a number between 1 and 9.")

        if not done(n):
            print("AI is thinking...")
            aiR,aiC=best(copy.deepcopy(n))
            n[aiR][aiC]=O
            print(f"AI played: Position {aiR*3+aiC+1}")
            board(n)

    if winner(n,X):
        print("Congratulations! You won!")
    elif winner(n,O):
        print("AI wins! Better luck next time!")
    else:
        print("It's a draw! Good game!")

```
